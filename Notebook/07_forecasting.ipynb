```python
# Predict future monthly revenue

import pandas as pd
import matplotlib.pyplot as plt

from sqlalchemy import create_engine

from sklearn.linear_model import LinearRegression
import numpy as np

engine = create_engine(
    "mysql+pymysql://root:@localhost:3309/ecommerce_analytics1"
)
print("Database Connected Successfully!")

```

```python
query = """
SELECT
    LEFT(o.order_purchase_timestamp,7) AS month,
    ROUND(SUM(p.payment_value),2) AS revenue
FROM orders o
JOIN payments p
ON o.order_id = p.order_id
GROUP BY month
ORDER BY month;
"""

revenue = pd.read_sql(query, engine)

revenue.head()
```

```python
revenue['month'] = pd.to_datetime(revenue['month'])

revenue = revenue.sort_values('month')

revenue['month_num'] = np.arange(len(revenue))

revenue.head()
```

```python
X = revenue[['month_num']]
y = revenue['revenue']

model = LinearRegression()

model.fit(X, y)

print("Model Trained Successfully!")
```
